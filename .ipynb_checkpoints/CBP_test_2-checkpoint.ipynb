```python
# Dependencies
import json
import requests
import pandas as pd
from config import api_key
```

```python
host = "http://api.census.gov/data"
year = "2018"
dataset = "cbp"
base_url = "/".join([host, year, dataset])
predicates = {}
get_vars = ["NAME","EMP"]
predicates["get"] = ",".join(get_vars)
predicates["for"] = "state:*"
```

```python
r = requests.get(base_url, params=predicates)
```

```python
r_json = r.json()[1:]
```

```python
r_df = pd.DataFrame(r_json)
r_df
```
